

**1. Modulo (`%`):**
- The modulo operation, denoted by `%`, returns the remainder when one number is divided by another.
- For two integers `a` and `b` (where `b` is not equal to zero), `a % b` returns the remainder of the division of `a` by `b`.
- It can also be applied to floating-point numbers, where it works in the same way but with floating-point results.
- For example, `10 % 3` returns `1` because the remainder of dividing 10 by 3 is 1.
- The result of the modulo operation is always non-negative or zero.

**2. Floor Division (`//` or integer division):**
- Floor division is the operation of dividing one number by another and rounding the result down to the nearest integer (towards negative infinity).
- In Python, floor division is denoted by `//`.
- For two integers `a` and `b` (where `b` is not equal to zero), `a // b` returns the largest integer that is less than or equal to `a` divided by `b`.
- For example, `10 // 3` returns `3` because the result of dividing 10 by 3 is approximately 3.333, and floor division rounds it down to the nearest integer, which is 3.

Here's a comparison between the two operations using an example:

- `10 / 3` (standard division) gives approximately `3.333`.
- `10 // 3` (floor division) gives `3` (rounded down to the nearest integer).
- `10 % 3` (modulo) gives `1` (the remainder when dividing 10 by 3).

In summary, floor division provides the quotient rounded down to the nearest integer, while modulo provides the remainder of a division operation. These operations are commonly used in programming for various tasks, such as splitting tasks into equal-sized portions, checking for even or odd numbers, and addressing cyclic behaviors in algorithms.

###String Indexing
In Python, strings are considered as sequences of characters. This means that Python assigns each character in a string a unique position, known as an index, which allows you to access specific parts of the string. Here are some key points to understand about string indexing:

1. Indexing starts from 0: In Python, indexing begins with 0. This means that the first character in a string is assigned an index of 0, the second character has an index of 1, and so on. This indexing scheme helps you pinpoint individual characters within a string.

2. Using Square Brackets: To access a specific character within a string, you can use square brackets `[]` after the string variable, followed by the index of the character you want to retrieve. For example, `my_string[0]` would access the first character of the string, and `my_string[1]` would access the second character.

3. Sequence of Numbers: The index of a character in a string is essentially a sequence of numbers that represent the character's position. This sequence starts with 0 and increases by 1 for each subsequent character.

4. Accessing Elements: By using these indices, you can access individual characters in the string, modify them, or perform various operations based on their positions.

Overall, string indexing is a fundamental concept in Python, allowing you to work with individual characters within a string by specifying their positions using integer indices.

````python
# Assign a string to a variable
string = 'Python'

# Access and print the first element (index 0)
first_element = string[0]
print("First Element:", first_element)  # Output: 'P'

# Access and print the second element (index 1)
second_element = string[1]
print("Second Element:", second_element)  # Output: 'y'

# Access and print the last element using negative index (-1)
last_element = string[-1]
print("Last Element:", last_element)  # Output: 'n'
````

###String Slicing:

String slicing in Python allows you to extract specific portions or substrings from a string. You can use the colon `:` to define slicing ranges. Here are the key points to understand about string slicing:

1. Basics of Slicing:
   - The syntax for slicing is `start_index:end_index`, where the `start_index` is inclusive (included in the result), and the `end_index` is exclusive (not included).
   - The element located at the `end_index` is not part of the sliced result.

2. Slicing with Starting and Ending Indices:
   - To extract a substring starting from a specific index and ending at another index, you can use slicing. For example, `string[2:5]` will extract characters from index 2 up to (but not including) index 5.

3. Missing Starting or Ending Indices:
   - If you omit the starting index, Python assumes you want to start from the beginning of the string. For example, `string[:5]` will extract characters from the beginning up to (but not including) index 5.
   - If you omit the ending index, Python assumes you want to slice all the way to the end. For example, `string[3:]` will extract characters from index 3 to the end of the string.

4. Extracting the Entire String:
   - To get the entire string, you can use a full colon `[:]` or omit both the starting and ending indices. For example, `string[:]` and `string` will both give you the entire string.

5. Negative Indexing:
   - Negative indices allow you to slice from the end of the string. For example, `string[-1]` refers to the last character, and `string[-4:]` extracts the last four characters.

6. Slicing with a Step:
   - You can specify a step size to skip characters while slicing. The default step size is 1. For example, `string[::2]` will extract every second character from the string.

7. Reversing a String:
   - You can use a step of -1 to reverse the string. For instance, `string[::-1]` will give you the string in reverse order.

8. Slicing with Negative Step:
   - Using a negative step allows you to reverse the sliced portion of the string. For example, `string[4:2:-1]` will give you the characters from index 4 to index 2 in reverse order.

Here's the original text rewritten with these points in mind:


````python
# Basics of Slicing
# Extract characters from index 2 to 5 (exclusive)
string[2:5]

# Missing Starting or Ending Indices
# Extract characters from the beginning up to index 5 (exclusive)
string[:5]

# Extract characters from index 3 to the end
string[3:]

# Extract the entire string
string[:]

# Negative Indexing
# Access the last character
string[-1]

# Extract the last four characters
string[-4:]

# Slicing with a Step
# Extract every second character
string[::2]

# Reversing a String
# Reverse the string
string[::-1]

# Slicing with Negative Step
# Extract characters from index 4 to 2 in reverse order
string[4:2:-1]
````

These examples should help illustrate the concepts of string slicing in Python.


### String Properties and Immutability

Strings in Python have an important property known as **immutability**. This means that once a string is created, the individual characters within it cannot be changed or replaced directly through item assignment. However, there are other ways to manipulate strings, as we'll explore.

### Attempting to Change a String

For example, if you try to change the first letter 'H' in the string 'Hello' to 'C' using item assignment, you will encounter an error:

```python
string = 'Hello'
string[0] = 'C'  # Error: 'str' object does not support item assignment
```

This error is telling us that you cannot change the contents of a string using item assignment. Strings are immutable, and individual characters cannot be reassigned directly.

### Concatenating Strings

While you cannot modify a string in place, you can create a new string by concatenating existing strings together. Here are some examples:

```python
string1 = 'abc'
string2 = 'def'

# Concatenate strings using the + operator
result = string1 + ' ' + string2
print(result)  # Output: 'abc def'

# Convert an integer to a string and concatenate
num = 4
result = string1 + str(num) + string2
print(result)  # Output: 'abc4def'
```

### Reassigning a String

If you want to change the value of a string, you can reassign it to a new value. This doesn't modify the original string but creates a new string with the updated value:

```python
string = 'Hello'
string = string + ' concatenate me!'
print(string)  # Output: 'Hello concatenate me!'
```

### Repetition

You can also repeat a string multiple times using the `*` operator:

```python
letters = 'wubba'
result = letters * 3
print(result)  # Output: 'wubbawubbawubba'
```

In summary, strings are immutable in Python, meaning you cannot change their contents directly. However, you can create new strings through concatenation, reassignment, and repetition to achieve the desired results.

### String Functions and Methods

In Python, strings are versatile data types that offer various functions and methods to perform different operations. Let's explore some of the essential string functions and methods.

1. **len() Function:**
   - The `len()` function returns the length of the string, which is the count of characters in the string.
   
   Example:
   ```python
   algorithm = 'Neural Networks'
   length = len(algorithm)
   print(length)  # Output: 14
   ```

2. **lower() Method:**
   - The `lower()` method converts a string to lowercase. It returns a new string with all characters in lowercase.
   
   Example:
   ```python
   algorithm = 'Neural Networks'
   lower_case = algorithm.lower()
   print(lower_case)  # Output: 'neural networks'
   ```

3. **upper() Method:**
   - The `upper()` method converts a string to uppercase. It returns a new string with all characters in uppercase.
   
   Example:
   ```python
   algorithm = 'Neural Networks'
   upper_case = algorithm.upper()
   print(upper_case)  # Output: 'NEURAL NETWORKS'
   ```

4. **count() Method:**
   - The `count()` method returns the count of a substring within the given string. It takes the substring as an argument.
   
   Example:
   ```python
   algorithm = 'Neural Networks'
   count_ne = algorithm.count('Ne')
   count_eu = algorithm.count('eu')
   print(count_ne)  # Output: 2
   print(count_eu)  # Output: 1
   ```

5. **find() Method:**
   - The `find()` method returns the index of the first occurrence of a substring within the given string. It takes the substring as an argument. If the substring is not found, it returns -1.
   
   Example:
   ```python
   algorithm = 'Neural Networks'
   index_e = algorithm.find('e')
   index_neural = algorithm.find('Neural')
   index_box = algorithm.find('Box')
   print(index_e)       # Output: 1
   print(index_neural)  # Output: 0
   print(index_box)     # Output: -1
   ```

6. **replace() Method:**
   - The `replace()` method replaces occurrences of a substring with another substring. It takes two arguments: the substring to replace and the replacement substring. It returns a modified string.
   
   Example:
   ```python
   algorithm = 'Neural Networks'
   replaced_spaces = algorithm.replace(' ', '-')
   replaced_n = algorithm.replace('N', 'L')
   print(replaced_spaces)  # Output: 'Neural-Networks'
   print(replaced_n)       # Output: 'Leural LeLworks'
   ```

7. **String Immunity:**
   - It's important to note that strings in Python are immutable, which means that you cannot change their contents directly. Methods like `lower()`, `upper()`, `count()`, `find()`, and `replace()` return modified strings, leaving the original string intact. To store changes, you need to assign them to a new variable.

8. **Formatted Strings (f-strings):**

##**Conditional Statements in Python**

```python
if condition:
```

This code will print "You are an adult." if the variable `age` is greater than or equal to 18, and "You are not an adult." otherwise.

We can also use the `elif` statement to check for multiple conditions. The `elif` statement has the following syntax:

```python

  # statements to execute if condition1 is true

  # statements to execute if condition2 is true

  # statements to execute if conditionN is true

  # statements to execute if none of the conditions are true
```

Here is an example of an `if-elif` statement:

```python
grade = "A"

if grade == "A":
  print("You got an A!")
elif grade == "B":
  print("You got a B.")

  print("You got a C.")

  print("You got a D or an F.")
```

This code will print the appropriate message based on the value of the variable `grade`.

We can also use nested conditional statements to create more complex logic. For example, the following code checks if a number is even and positive:

```python
number = 10

if number % 2 == 0:
  if number > 0:
    print("The number is even and positive.")
  else:
    print("The number is even and negative.")

  print("The number is odd.")
```

**Truthy and Falsy Values**

In Python, all values are considered **truthy** except for the following, which are **falsy**:

* `None`
* `False`
* `0`
* `0.0`
* `[]` - an empty list
* `{}` - an empty dictionary
* `()` - an empty tuple
* `''` - an empty string

This means that `if` statements will evaluate to `True` if the condition is any truthy value, and to `False` if the condition is falsy.

The first `if` statement will evaluate to `False` because the condition is an empty list, which is a falsy value. The second `if` statement will evaluate to `True` because the condition is 1, which is a truthy value.
